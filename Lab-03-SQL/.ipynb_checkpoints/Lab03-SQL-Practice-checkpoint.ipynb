- Đọc kỹ yêu cầu trước khi viết code.


```python
import sqlite3
import pandas as pd

# Tạo hoặc kết nối tới database
conn = sqlite3.connect('smartlib.db')
cursor = conn.cursor()

def run_sql(sql):
    """Hàm chạy SQL và hiển thị kết quả dạng bảng"""
    try:
        df = pd.read_sql_query(sql, conn)
        display(df)
    except Exception as e:
        print(f"LỖI SQL: {e}")

print("Kết nối thành công đến SmartLib Database!")
```

## PHẦN 1: PHÂN TÍCH VÀ THIẾT KẾ CƠ SỞ DỮ LIỆU (DDL)
Hệ thống Thư viện SmartLib cần quản lý 4 thực thể chính:

- TacGia (Authors): Mã tác giả (Khóa chính), Tên tác giả, Quốc tịch.
- Sach (Books): Mã sách (Khóa chính), Tên sách, Năm xuất bản, Số lượng tồn kho, Mã tác giả (Khóa ngoại).
- DocGia (Members): Mã đọc giả (Khóa chính), Tên đọc giả, Email, Năm tham gia.
- MuonTra (BorrowRecords): Mã phiếu mượn (Khóa chính), Mã sách (Khóa ngoại), Mã đọc giả (Khóa ngoại), Ngày mượn, Ngày trả (Nếu chưa trả thì để NULL).

#### NHIỆM VỤ CỦA BẠN:
Viết câu lệnh SQL để tạo 4 bảng trên. Yêu cầu:

- Kiểu dữ liệu phải phù hợp (TEXT cho chuỗi, INTEGER cho số, v.v.).
- Bắt buộc có ràng buộc PRIMARY KEY cho các mã.
- Bắt buộc có ràng buộc FOREIGN KEY cho các mã liên kết.
- Cột TenSach, TenTacGia, TenDocGia, Email không được phép để trống (NOT NULL).

## PHẦN 3: TRUY VẤN DỮ LIỆU CƠ BẢN (DQL)
Sử dụng các câu lệnh SELECT, WHERE, ORDER BY, LIMIT để thực hiện các yêu cầu sau. Mỗi yêu cầu viết trong 1 cell riêng biệt.

- Yêu cầu 3.1: Tìm tất cả các cuốn sách được xuất bản sau năm 1950.
- Yêu cầu 3.2: Tìm tên của tất cả đọc giả có email chứa chuỗi "@email.com", sắp xếp theo tên đọc giả tăng dần (A-Z).
- Yêu cầu 3.3: Liệt kê top 2 cuốn sách có số lượng tồn kho cao nhất.

## PHẦN 4: THỐNG KÊ DỮ LIỆU (AGGREGATION & GROUP BY)
- Yêu cầu 4.1: Thống kê tổng số lượng sách hiện có trong thư viện (Tổng của cột TồnKho trong bảng Sach).
- Yêu cầu 4.2: Đếm xem mỗi tác giả có bao nhiêu cuốn sách trong thư viện. (Hiển thị Tên tác giả và Số lượng sách).
- Yêu cầu 4.3: Tính số lượng đọc giả tham gia theo từng năm. (Nhóm theo cột Năm tham gia).

## PHẦN 5: KẾT NỐI BẢNG (JOINS) - PHẦN QUAN TRỌNG NHẤT
- Yêu cầu 5.1 (INNER JOIN): Hiển thị danh sách  Tên sách và Tên tác giả tương ứng của tất cả các sách có trong thư viện
- Yêu cầu 5.2 (2 TABLES JOIN): Hiển thị Tên đọc giả và Tên sách mà họ đã mượn. (Chỉ hiển thị những người đã có phiếu mượn).
- Yêu cầu 5.3 (MULTI-JOIN): Hiển thị đầy đủ thông tin 1 phiếu mượn bao gồm: Tên Đọc giả, Tên Sách, Tên Tác giả, Ngày mượn, Ngày trả.
- Yêu cầu 5.4 (LEFT JOIN): Thư viện muốn biết TẤT CẢ các cuốn sách, và nếu cuốn sách đó đang có người mượn (chưa trả) thì hiện tên người mượn.
(Gợi ý: JOIN từ bảng Sach sang MuonTra với điều kiện Ngày trả IS NULL. Sách nào không có ai mượn sẽ hiện NULL ở cột tên người mượn).

## PHẦN 6: TRUY VẤN CON (SUBQUERY) & CẬP NHẬT DỮ LIỆU
- Yêu cầu 6.1 (SUBQUERY trong WHERE): Tìm ra tên của những cuốn sách có số lượng tồn kho lớn hơn mức tồn kho trung bình của toàn thư viện.
- Yêu cầu 6.2 (UPDATE): Cuốn "Harry Potter 1" vừa được thư viện nhập thêm stock. Hãy viết lệnh cập nhật (UPDATE) tăng số lượng tồn kho của cuốn sách này lên thêm 5 cuốn. Sau đó chạy lệnh SELECT để kiểm tra lại kết quả.
- Yêu cầu 6.3 (DELETE với Subquery): Xóa tất cả các Đọc giả chưa từng mượn cuốn sách nào khỏi bảng DocGia.
(Gợi ý: Sử dụng DELETE ... WHERE MaDocGia NOT IN (SELECT ... FROM MuonTra))
