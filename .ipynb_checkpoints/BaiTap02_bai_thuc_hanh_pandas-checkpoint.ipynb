### 4.2 Xu ly gia tri null

Co 3 chien luoc chinh:
- **Xoa dong** (`dropna`): dung khi ty le null thap va dong do khong quan trong
- **Dien gia tri** (`fillna`): dien trung binh, trung vi, gia tri truoc/sau do, hoac hang mac dinh
- **Noi suy** (`interpolate`): cho du lieu chuoi thoi gian

```python
# Tao ban sao de lam sach (khong chinh sua du lieu goc)
df_clean = df_loi.copy()

print('Truoc khi xu ly null:')
print(df_clean.isnull().sum())
```

```python
# Xoa dong co null o cot Ho_ten (truong bat buoc)
df_clean = df_clean.dropna(subset=['Ho_ten'])
print(f'Sau khi xoa dong null Ho_ten: {len(df_clean)} dong')

# Dien null o cot Tuoi bang trung vi
trung_vi_tuoi = df_clean['Tuoi'].median()
df_clean['Tuoi'] = df_clean['Tuoi'].fillna(trung_vi_tuoi)
print(f'Dien Tuoi null bang trung vi: {trung_vi_tuoi}')

# Dien null o cot Luong bang trung binh
trung_binh_luong = df_clean['Luong'].mean()
df_clean['Luong'] = df_clean['Luong'].fillna(trung_binh_luong)
print(f'Dien Luong null bang trung binh: {trung_binh_luong:,.0f}')

# Dien null Ngay_sinh va Thanh_pho bang 'Chua ro'
df_clean['Ngay_sinh'] = df_clean['Ngay_sinh'].fillna('Chua ro')

print('\nSau khi xu ly null:')
print(df_clean.isnull().sum())
```

### 4.3 Xu ly gia tri ngoai le (Outlier)

Gia tri ngoai le la gia tri bat thuong, qua lon hoac qua nho so voi phan con lai.
Phuong phap phat hien:
- **Z-score**: gia tri > 3 lan do lech chuan
- **IQR**: gia tri ngoai khoang Q1 - 1.5*IQR den Q3 + 1.5*IQR

```python
# Phat hien ngoai le o cot Tuoi bang IQR
Q1 = df_clean['Tuoi'].quantile(0.25)
Q3 = df_clean['Tuoi'].quantile(0.75)
IQR = Q3 - Q1
gioi_han_tren = Q3 + 1.5 * IQR
gioi_han_duoi = Q1 - 1.5 * IQR

print(f'Q1={Q1}, Q3={Q3}, IQR={IQR}')
print(f'Gioi han hop le: [{gioi_han_duoi:.0f}, {gioi_han_tren:.0f}]')

outlier_tuoi = df_clean[~df_clean['Tuoi'].between(gioi_han_duoi, gioi_han_tren)]
print(f'\nCac dong co Tuoi ngoai le:')
print(outlier_tuoi[['STT','Ho_ten','Tuoi']])

# Xu ly: thay bang gia tri trung vi
df_clean.loc[~df_clean['Tuoi'].between(gioi_han_duoi, gioi_han_tren), 'Tuoi'] = trung_vi_tuoi
print(f'\nDa thay the gia tri ngoai le bang trung vi ({trung_vi_tuoi})')
```

```python
# Phat hien va xu ly luong <= 0
print('Dong co luong <= 0:')
print(df_clean[df_clean['Luong'] <= 0][['STT','Ho_ten','Luong']])

df_clean.loc[df_clean['Luong'] <= 0, 'Luong'] = trung_binh_luong
print('Da thay the luong <= 0 bang trung binh')
```

## PHAN 5: Tao them Cot va Bien doi Du lieu

### 5.1 Tao cot moi bang phep tinh toan

```python
# Tao them cot tinh toan tu bang nhan_vien
df_nv2 = df_nv.copy()

# Tuoi nhan vien
df_nv2['Tuoi'] = 2025 - df_nv2['Nam_sinh']

# Tong thu nhap (luong + phu cap)
df_nv2['Tong_thu_nhap'] = df_nv2['Luong_co_ban'] + df_nv2['Phu_cap']

# Ty le thue tren luong co ban
df_nv2['Ty_le_thue_pct'] = (df_nv2['Thue_TNCN'] / df_nv2['Luong_co_ban'] * 100).round(2)

# Xep loai luong
def xep_loai_luong(luong):
    if luong < 10_000_000:
        return 'Thap'
    elif luong < 20_000_000:
        return 'Trung binh'
    elif luong < 30_000_000:
        return 'Kha'
    else:
        return 'Cao'

df_nv2['Muc_luong'] = df_nv2['Luong_co_ban'].apply(xep_loai_luong)

# Nhom tuoi
df_nv2['Nhom_tuoi'] = pd.cut(
    df_nv2['Tuoi'],
    bins=[0, 30, 40, 50, 100],
    labels=['Duoi 30', '30-40', '40-50', 'Tren 50']
)

cols_xem = ['MaNV','Ho_dem','Ten','Tuoi','Nhom_tuoi','Luong_co_ban','Muc_luong',
           'Tong_thu_nhap','Ty_le_thue_pct']
df_nv2[cols_xem].head(10)
```

### 5.2 np.where va np.select — Tao cot co dieu kien

`np.where(dieu_kien, gia_tri_neu_dung, gia_tri_neu_sai)` la cach nhanh hon viet ham.

```python
# Cot co_uu_dai: nhan vien danh gia Xuat sac hoac co luong thap duoi 12M
df_nv2['Co_uu_dai'] = np.where(
    (df_nv2['Danh_gia'] == 'Xuat sac') | (df_nv2['Luong_co_ban'] < 12_000_000),
    'Co',
    'Khong'
)

# np.select: nhieu dieu kien tren nhieu lua chon (giong if-elif-else)
conditions = [
    df_nv2['Danh_gia'] == 'Xuat sac',
    df_nv2['Danh_gia'] == 'Tot',
    df_nv2['Danh_gia'] == 'Kha',
    df_nv2['Danh_gia'] == 'Trung binh',
]
choices = [1.15, 1.10, 1.05, 1.02]  # He so tang luong

df_nv2['He_so_tang'] = np.select(conditions, choices, default=1.0)
df_nv2['Luong_moi']  = (df_nv2['Luong_co_ban'] * df_nv2['He_so_tang']).round(-4)

df_nv2[['MaNV','Danh_gia','Luong_co_ban','He_so_tang','Luong_moi']].head(10)
```

### 5.3 .apply() — Ap dung ham tu dinh nghia len moi dong/cot

```python
# apply voi axis=1 chay tren tung dong
def tinh_thuong(row):
    """
    Tinh thuong cuoi nam dua tren danh gia va so ngay phep con lai.
    Quy tac: Xuat sac=3 thang luong, Tot=2, Kha=1.5, TB=1, Yeu=0
    Giam 5% cho moi ngay phep da su dung
    """
    he_so_map = {'Xuat sac': 3.0, 'Tot': 2.0, 'Kha': 1.5, 'Trung binh': 1.0, 'Yeu': 0.0}
    he_so = he_so_map.get(row['Danh_gia'], 0)
    ngay_da_su_dung = 12 - row['So_ngay_phep']
    he_so_thuc = he_so * max(0, 1 - ngay_da_su_dung * 0.05)
    return round(row['Luong_co_ban'] * he_so_thuc, -4)

df_nv2['Thuong_cuoi_nam'] = df_nv2.apply(tinh_thuong, axis=1)

df_nv2[['MaNV','Danh_gia','So_ngay_phep','Luong_co_ban','Thuong_cuoi_nam']].head(10)
```

## PHAN 7: Thong ke va Phan tich Theo Nhom — groupby

### 7.1 groupby co ban

`groupby` hoat dong theo 3 buoc: **Split** (phan nhom) -> **Apply** (tinh toan) -> **Combine** (tong hop ket qua)

```python
# Thong ke co ban theo phong ban
thong_ke_pb = df_nv.groupby('Phong_ban')['Luong_co_ban'].agg(
    so_nguoi='count',
    luong_tb='mean',
    luong_min='min',
    luong_max='max',
    luong_tong='sum',
    do_lech_chuan='std'
).round(0)

thong_ke_pb.sort_values('luong_tb', ascending=False)
```

```python
# Thong ke nhieu cot cung luc
thong_ke_da_chieu = df_nv.groupby('Phong_ban').agg(
    so_nv=('MaNV', 'count'),
    luong_tb=('Luong_co_ban', 'mean'),
    phu_cap_tb=('Phu_cap', 'mean'),
    luong_thuc_tb=('Luong_thuc_lanh', 'mean'),
    ngay_phep_tb=('So_ngay_phep', 'mean')
).round(0)

thong_ke_da_chieu
```

```python
# Phan tich theo 2 chieu: phong ban + gioi tinh
phan_tich_2c = df_nv.groupby(['Phong_ban', 'Gioi_tinh']).agg(
    so_nguoi=('MaNV', 'count'),
    luong_tb=('Luong_co_ban', 'mean')
).round(0)

phan_tich_2c
```

```python
# Phan tich don hang theo khu vuc va trang thai
phan_tich_dh = df_dh.groupby(['Khu_vuc', 'Trang_thai']).agg(
    so_don=('Ma_don', 'count'),
    tong_doanh_thu=('Thanh_tien', 'sum'),
    trung_binh_don=('Thanh_tien', 'mean')
).round(0)

phan_tich_dh
```

### 7.2 transform — Giu nguyen kich thuoc DataFrame

Khac voi `agg` (giam so dong), `transform` tra ve Series cung kich thuoc ban dau.
Dung de tao cot moi co gia tri tinh theo nhom.

```python
# Luong trung binh cua phong ban cho moi nhan vien
df_nv2['Luong_tb_phong'] = df_nv2.groupby('Phong_ban')['Luong_co_ban'].transform('mean').round(0)

# So sanh luong ca nhan voi trung binh phong ban
df_nv2['Chech_lech_luong'] = df_nv2['Luong_co_ban'] - df_nv2['Luong_tb_phong']
df_nv2['Chech_lech_pct']   = (df_nv2['Chech_lech_luong'] / df_nv2['Luong_tb_phong'] * 100).round(2)

cols = ['MaNV','Ho_dem','Ten','Phong_ban','Luong_co_ban','Luong_tb_phong',
        'Chech_lech_luong','Chech_lech_pct']
df_nv2[cols].sort_values('Chech_lech_pct', ascending=False).head(10)
```

## PHAN 8: Ket nop Du lieu — merge, join, concat

### 8.1 merge — tuong tu SQL JOIN

| Kieu | Y nghia |
|------|---------|
| `inner` | Chi giu cac hang co khoa ton tai o ca 2 bang |
| `left`  | Giu tat ca hang bang trai, NaN neu khong khop |
| `right` | Giu tat ca hang bang phai |
| `outer` | Giu tat ca hang ca 2 bang |

```python
# Ket nop don hang voi thong tin nhan vien
# df_dh co cot Ma_NV, df_nv co cot MaNV
df_dh_nv = pd.merge(
    df_dh,
    df_nv[['MaNV', 'Ho_dem', 'Ten', 'Phong_ban', 'Tinh_thanh']],
    left_on='Ma_NV',
    right_on='MaNV',
    how='left'
)

print(f'df_dh ban dau: {df_dh.shape}')
print(f'Sau khi merge: {df_dh_nv.shape}')
print(f'Don hang chua tim thay NV: {df_dh_nv["Ho_dem"].isnull().sum()}')

df_dh_nv[['Ma_don','Ngay_dat','Ma_NV','Ho_dem','Ten','Phong_ban','Ten_SP','Thanh_tien']].head(8)
```

```python
# Kiem tra don hang co nhan vien nao ban chay nhat
doanh_so_nv = (
    df_dh_nv
    .groupby(['Ma_NV', 'Ho_dem', 'Ten', 'Phong_ban'])
    .agg(
        so_don=('Ma_don', 'count'),
        tong_doanh_thu=('Thanh_tien', 'sum')
    )
    .reset_index()
    .sort_values('tong_doanh_thu', ascending=False)
)
print('Top 10 nhan vien doanh so cao nhat:')
doanh_so_nv.head(10)
```

## PHAN 9: Pivot Table va Phan tich Cheo

### 9.1 pivot_table

Pivot table la cong cu manh de tom tat du lieu theo 2 chieu, tuong tu PivotTable trong Excel.

```python
# Pivot: luong trung binh theo phong ban va trinh do hoc van
pivot_luong = pd.pivot_table(
    df_nv,
    values='Luong_co_ban',
    index='Phong_ban',
    columns='Trinh_do',
    aggfunc='mean',
    fill_value=0
).round(0)

print('Luong trung binh (dong = Phong ban, cot = Trinh do):')
pivot_luong
```

```python
# Pivot: so nhan vien theo phong ban va danh gia
pivot_danh_gia = pd.pivot_table(
    df_nv,
    values='MaNV',
    index='Phong_ban',
    columns='Danh_gia',
    aggfunc='count',
    fill_value=0
)

# Them cot tong
pivot_danh_gia['TONG'] = pivot_danh_gia.sum(axis=1)
print('So nhan vien theo phong ban va danh gia:')
pivot_danh_gia
```

```python
# Truc quan hoa pivot bang heatmap
fig, ax = plt.subplots(figsize=(10, 5))
pivot_num = pivot_luong.replace(0, float('nan'))
im = ax.imshow(pivot_num.values, cmap='YlOrRd', aspect='auto')
plt.colorbar(im, ax=ax, label='Luong co ban (VND)')

ax.set_xticks(range(len(pivot_num.columns)))
ax.set_xticklabels(pivot_num.columns, rotation=30, ha='right')
ax.set_yticks(range(len(pivot_num.index)))
ax.set_yticklabels(pivot_num.index)
ax.set_title('Luong trung binh theo Phong ban va Trinh do')

for i in range(len(pivot_num.index)):
    for j in range(len(pivot_num.columns)):
        val = pivot_num.values[i, j]
        if not np.isnan(val):
            ax.text(j, i, f'{val/1e6:.1f}M', ha='center', va='center', fontsize=8)

plt.tight_layout()
plt.show()
```

```python
# Pivot don hang: tong doanh thu theo khu vuc va danh muc
pivot_doanh_thu = pd.pivot_table(
    df_dh,
    values='Thanh_tien',
    index='Khu_vuc',
    columns='Danh_muc',
    aggfunc='sum',
    fill_value=0,
    margins=True,          # Them dong/cot tong cong
    margins_name='TONG'
).round(0)

print('Tong doanh thu theo Khu vuc va Danh muc:')
pivot_doanh_thu
```

## PHAN 12: Bai Toan Tong Hop — Bao cao Nhan su Hang Quy

Yeu cau: Xay dung mot bao cao nhan su day du bao gom:
- Tong hop theo phong ban
- Danh sach nhan vien can tang luong (danh gia tot, luong duoi trung binh phong ban)
- Phan tich chi phi luong toan cong ty
- Xuat ra file Excel nhieu sheet co dinh dang

```python
# ---- BUOC 1: Chuan bi du lieu ----
df_bao_cao = df_nv.copy()
df_bao_cao['Tuoi'] = 2025 - df_bao_cao['Nam_sinh']
df_bao_cao['Luong_tb_phong'] = (
    df_bao_cao.groupby('Phong_ban')['Luong_co_ban'].transform('mean').round(0)
)
df_bao_cao['Duoi_trung_binh'] = df_bao_cao['Luong_co_ban'] < df_bao_cao['Luong_tb_phong']

print('Chuan bi du lieu hoan tat')
```

```python
# ---- BUOC 2: Tong hop phong ban ----
tong_hop_pb = df_bao_cao.groupby('Phong_ban').agg(
    So_NV             =('MaNV', 'count'),
    NV_Nam            =('Gioi_tinh', lambda x: (x=='Nam').sum()),
    NV_Nu             =('Gioi_tinh', lambda x: (x=='Nu').sum()),
    Luong_TB          =('Luong_co_ban', 'mean'),
    Luong_Min         =('Luong_co_ban', 'min'),
    Luong_Max         =('Luong_co_ban', 'max'),
    Tong_Chi_Phi_Luong=('Luong_thuc_lanh', 'sum'),
    NV_Xuat_Sac       =('Danh_gia', lambda x: (x=='Xuat sac').sum()),
    NV_Yeu            =('Danh_gia', lambda x: (x=='Yeu').sum()),
    Tuoi_TB           =('Tuoi', 'mean')
).round(0)

print('=== TONG HOP THEO PHONG BAN ===')
tong_hop_pb
```

```python
# ---- BUOC 3: Danh sach nhan vien can xem xet tang luong ----
can_tang_luong = df_bao_cao[
    df_bao_cao['Danh_gia'].isin(['Xuat sac', 'Tot']) &
    df_bao_cao['Duoi_trung_binh']
][['MaNV','Ho_dem','Ten','Phong_ban','Chuc_vu','Danh_gia',
   'Luong_co_ban','Luong_tb_phong']].copy()

can_tang_luong['Muc_de_xuat_tang'] = (
    (can_tang_luong['Luong_tb_phong'] - can_tang_luong['Luong_co_ban']) * 0.8
).round(-4)

can_tang_luong = can_tang_luong.sort_values('Muc_de_xuat_tang', ascending=False)

print(f'So nhan vien de xuat tang luong: {len(can_tang_luong)}')
can_tang_luong
```

```python
# ---- BUOC 4: Phan tich chi phi luong ----
tong_chi_phi = df_bao_cao['Luong_thuc_lanh'].sum()
chi_phi_theo_pb = (
    df_bao_cao.groupby('Phong_ban')['Luong_thuc_lanh']
    .sum()
    .sort_values(ascending=False)
)

print(f'Tong chi phi luong/thang: {tong_chi_phi:,.0f} VND')
print(f'Tong chi phi luong/nam  : {tong_chi_phi*12:,.0f} VND')
print()
for pb, cp in chi_phi_theo_pb.items():
    pct = cp / tong_chi_phi * 100
    print(f'  {pb:<15}: {cp:>15,.0f} VND  ({pct:.1f}%)')
```

```python
# ---- BUOC 5: Xuat bao cao ra Excel ----
with pd.ExcelWriter('bao_cao_nhan_su_Q4_2025.xlsx', engine='openpyxl') as writer:
    tong_hop_pb.reset_index().to_excel(
        writer, sheet_name='1_Tong_hop_Phong_ban', index=False)

    can_tang_luong.to_excel(
        writer, sheet_name='2_De_xuat_Tang_luong', index=False)

    df_bao_cao.sort_values(['Phong_ban','Luong_co_ban'], ascending=[True,False]).to_excel(
        writer, sheet_name='3_Chi_tiet_Nhan_vien', index=False)

    chi_phi_theo_pb.reset_index().rename(
        columns={'Luong_thuc_lanh':'Chi_phi_thang'}
    ).to_excel(writer, sheet_name='4_Chi_phi_Luong', index=False)

print('Da xuat bao_cao_nhan_su_Q4_2025.xlsx')
print('4 sheets: Tong_hop_Phong_ban | De_xuat_Tang_luong | Chi_tiet_Nhan_vien | Chi_phi_Luong')
```
